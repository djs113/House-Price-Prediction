```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns

df = pd.read_csv("Bengaluru_House_Data.csv")
df
```

```python
columns_of_interest = ['area_type', 'size', 'bath', 'balcony']

for column in columns_of_interest:
    plt.figure(figsize=(12, 6))
    sns.countplot(x=column, data=df, order=df[column].value_counts().index)
    plt.xticks(rotation=45, ha='right')
    plt.title(f'Bar Graph Analysis of {column}')
    plt.xlabel(column)
    plt.ylabel('Count')
    plt.tight_layout()
    plt.savefig(f'{column}_bar_graph.png')
    plt.show()
    
```

```python
total_sqft = df['total_sqft']
price = df['price']

plt.figure(figsize=(10, 6))
plt.scatter(total_sqft, price, alpha=0.5, color='r')
plt.title('Scatter Plot of Total Sqft vs. Price')
plt.xlabel('Total Sqft')
plt.ylabel('Price')

plt.show()
```
